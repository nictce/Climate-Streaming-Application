```python
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import pandas as pd
import pymongo
from pprint import pprint
import datetime as dt
```

```python
def publish(producer, topic, data):
    try:
        byte_data = bytes(data, encoding = 'utf-8')
        producer.send(topic, value = byte_data)
#         producer.flush()
        print('Message sent successfully!')
    except Exception as e:
        print(f'Exception occured while publishing message: {str(e)}')

def connect_kafka_producer():
    producer = None
    try:
        producer = KafkaProducer(bootstrap_servers=['127.0.0.1:9092'], api_version=(0, 10))
    except Exception as e:
        print(f'Exception occured while connecting to Kafka: {str(e)}')
    finally:
        return producer
```

```python
def find_latest_date():
    client = pymongo.MongoClient()
    collection = client.fit_3182_assignment_db.historic

    res = collection.find().sort('date', -1).limit(1)
    latest_date = next(res, None)['date']
    return latest_date

def terra_csv_to_dict():
    terra_df = pd.read_csv('hotspot_TERRA_streaming.csv')
    terra_streaming = []

    for _, row in terra_df.iterrows():
        doc = {}
        doc['latitude'] = float(row['latitude'])
        doc['longitude'] = float(row['longitude'])
        doc['confidence'] = float(row['confidence'])
        doc['surface_temperature_celcius'] = float(row['surface_temperature_celcius'])

        terra_streaming.append(doc)
    
    return terra_streaming
    
```

```python
if __name__ == '__main__':

    latest_date = find_latest_date()
    terra_streaming = terra_csv_to_dict()
    nrows = len(terra_streaming)

    topic = 'Terra'
    print('Publishing records..')
    climate_producer = connect_kafka_producer()

    while True:
        latest_date += dt.timedelta(hours = 4, minutes = 48)
        randrow = terra_streaming[random.randrange(nrows)]
        randrow['date'] = str(latest_date)#.isoformat()
        randrow['station'] = 'terra_producer'
        publish(climate_producer, topic, dumps(randrow))
        sleep(2)
```
